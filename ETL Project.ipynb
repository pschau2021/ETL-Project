```python
import pandas as pd 

df1 = "etl/chicagocrime.csv"

df2 = "etl/lacrime.csv"

chicago_data = pd.read_csv(df1)

la_data = pd.read_csv(df2)

colsd = [3,6,7,8,9,10,11,12,13,16]

chicago_data = chicago_data.drop(chicago_data.columns[colsd], axis =1)

cols = [0,1,2,3,4,5,6]

chicago_data["DESCRIPTION"] = chicago_data[" PRIMARY DESCRIPTION"] + chicago_data[" SECONDARY DESCRIPTION"]

colsd2 = [3,4]

chicago_data = chicago_data.drop(chicago_data.columns[colsd2], axis =1)

chicago_data.rename(columns={'BLOCK': 'LOCATION'}, inplace=True)
chicago_data.head()
```

```python
colsd1 = [1,3,4,5,6,7,8,10,11,12,13,14,15,16,17,19]

la_data = la_data.drop(la_data.columns[colsd1], axis =1)

la_data.rename(columns={'DR_NO': 'CASE#', 'DATE OCC': 'DATE  OF OCCURRENCE','Crm Cd Desc': 'DESCRIPTION','LAT': 'LATITUDE','LON': 'LONGITUDE' }, inplace=True)

la_data.head()
```

```python
colsd1 = [3,4,5,6,7,9]

la_data = la_data.drop(la_data.columns[colsd1], axis =1)

la_data.head()
```
